```python
for column in Car.columns:
  if Car[column].isna().any():
    if Car[column].dtype == 'int64':
      Car[column] = Car[column].fillna(Car[column].mode())
    if Car[column].dtype == 'float64':
      Car[column] = Car[column].fillna(Car[column].mean())
```
